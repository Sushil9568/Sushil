```python
import tkinter as tk

# Function for handling button click events
def click_button(value):
    current = entry.get()  # Get current value in the entry widget
    entry.delete(0, tk.END)  # Clear the entry widget
    entry.insert(tk.END, current + value)  # Append the clicked value

# Function for performing the calculation
def calculate():
    try:
        result = eval(entry.get())  # Use eval to evaluate the expression in the entry
        entry.delete(0, tk.END)
        entry.insert(tk.END, result)  # Display the result in the entry
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")  # In case of error, display "Error"

# Function for clearing the entry
def clear():
    entry.delete(0, tk.END)  # Clear the entry widget

# Function for handling the backspace (delete one character)
def backspace():
    current = entry.get()
    entry.delete(len(current) - 1, tk.END)  # Delete the last character

# Create the main window
window = tk.Tk()
window.title("Calculator")

# Create an entry widget to display the current input or result
entry = tk.Entry(window, width=20, font=("Arial", 24), borderwidth=2, relief="solid", justify="right")
entry.grid(row=0, column=0, columnspan=4)

# Create buttons for numbers 0-9 and other operations
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2),
    ('0', 4, 1), ('+', 1, 3), ('-', 2, 3),
    ('*', 3, 3), ('/', 4, 3), ('=', 4, 2),
    ('C', 5, 0), ('Backspace', 5, 1)
]

# Add the buttons to the window
for (text, row, col) in buttons:
    if text == "=":
        button = tk.Button(window, text=text, width=5, height=2, font=("Arial", 18), command=calculate)
    elif text == "C":
        button = tk.Button(window, text=text, width=5, height=2, font=("Arial", 18), command=clear)
    elif text == "Backspace":
        button = tk.Button(window, text=text, width=10, height=2, font=("Arial", 18), command=backspace)
    else:
        button = tk.Button(window, text=text, width=5, height=2, font=("Arial", 18), command=lambda value=text: click_button(value))
    
    button.grid(row=row, column=col, padx=5, pady=5)

# Start the Tkinter event loop
window.mainloop()
```
